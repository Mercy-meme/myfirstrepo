### CRUD on Txt File

```python
#READ the contents of the file
with open ("TextFile.txt", 'r') as file:
    content=file.readlines()
    
print("The following is what is in the txt file:")
for i in content:
    print(i)
```

```python
#READ the contents of the file
with open ("TextFile.txt", 'r') as file:
    content=file.readlines()
    
print("The following is what is in the txt file:")
for i in content:
    print(i.strip())
```

```python
#Update an existing record in a txt file

content[2] = "Joanna 9\n"
with open("TextFile.txt", 'w') as file:
    file.writelines(content)
    
print("File successfully updated.")
for i in content:
    print(i.strip())
    
```
