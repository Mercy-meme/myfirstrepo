```python
#Concat along rows uses hstack
import numpy as np

arr1 = np.array([1, 2, 3]) #d

arr2 = np.array([4, 5, 6])

arr = np.hstack((arr1, arr2))

print(arr)
```

```python
#Concat along columns ()uses vstack--depends on the nature of the dataset

arr1 = np.array([1, 2, 3]) #df

arr2 = np.array([4, 5, 6])

arr = np.vstack((arr1, arr2))

print(arr)
```
